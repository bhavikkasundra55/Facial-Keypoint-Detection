### Reading and organising data

```python
import os
```

```python
data = pd.read_csv('F:/Github/Facial-Keypoint-Detection/data/training_frames_keypoints.csv')
data.head()
```

```python
images = []

for index in range(50):

    image_name = data.iloc[index,0]
    image = mpimg.imread(os.path.join('F:/Github/Facial-Keypoint-Detection/data/training', image_name))
    cropped_img = tf.image.resize(image,size=[128,128])
    images.append(cropped_img)

print(images)
```
